```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df = df[df['TotalCharges'].notnull()]
```

```python
df['Churn'] = df['Churn'].map({'Yes':1, 'No':0})
```

```python
df.shape
```

```python
df['Churn'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")   # add title
plt.savefig("contract_churn.png", bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.savefig("monthly_charges.png", bbox_inches='tight')
plt.show()
```

```python
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Churn by Internet Service")
plt.savefig("internet_service.png", bbox_inches='tight')
plt.show()
```

```python
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title("Churn by Payment Method")
plt.xticks(rotation=30)
plt.savefig("payment_method.png", bbox_inches='tight')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df[df['TotalCharges'].notnull()]
df['Churn'] = df['Churn'].map({'Yes':1, 'No':0})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x='tenure', hue='Churn', bins=30)
plt.title("Churn vs Tenure")
plt.show()
```
